Давайте попробуем перевести картинку из RGB в Lab!     
Для этого воспользуемся методами rgb2lab, lab2rgb класса color библиотеки skimage.

```python
from skimage.io import imread, imsave
from skimage.color import rgb2lab, lab2rgb
import matplotlib.pylab as plt
import numpy as np
```

```python
im = imread('images/hqdefault.jpg')
im1 = rgb2lab(im)
```

Отлично! Мы получили картинку в Lab-формате. Давайте выведем на экран изображение в сером цвете. Для этого обнулим второй и третий каналы изображения в Lab:

```python
im1[...,1] = im1[...,2] = 0
```

Получим изображение в градациях серого, преобразовав изображение обратно в цветовое пространство RGB из цветового пространства Lab:

```python
im_grey = lab2rgb(im1)
```

Посмотрим, что получилось:

```python
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'), plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(im_grey), plt.axis('off'), plt.title('Gray scale image', size=20)
plt.show()
```

# ЗАДАНИЕ 2.3
- Измените яркость исходного изображения, увеличив значение первого канала на 30.
- Уберите из исходного изображения сине-жёлтые тона.

# Решение

```python
# загружаем исходное изображение
im = imread('images/hqdefault.jpg')
# переводим в Lab
im1 = rgb2lab(im)

im1[...,0] += 30 # увеличиваем значение 1 канала на 30
im1[...,2] = 0 # зануляем 3 канал, т.е. убираем сине-желтые тона

# переводим обратно в rgb
im_final = lab2rgb(im1)

# сохраняем изображение
imsave('images/final.jpg', (im_final*255).astype('uint8'))
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'), plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(im_final), plt.axis('off'), plt.title('Final image', size=20)
plt.show()
```
